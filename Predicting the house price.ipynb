```python
import statsmodels.api as sm
X=np.append(arr=np.ones((1021,1)).astype(int),values=X,axis=1)
```

```python
high=1.0
feat=np.arange(249)
while high>0.045:
    X_opt=X[:,feat]
    regressor_Ols=sm.OLS(endog=y,exog=X_opt).fit()
    results_summary=regressor_Ols.summary()
    # Note that tables is a list. The table at index 1 is the "core" table. Additionally, read_html puts dfs in a list, so we want index 0
    results_as_html = results_summary.tables[1].as_html()
    fin_res=pd.read_html(results_as_html, header=0, index_col=0)[0]
    fin_res=fin_res.drop(columns=['coef','std err','t','[0.025','0.975]'])
    high=fin_res.loc[fin_res['P>|t|'].idxmax()][0]
    fin_res=fin_res.reset_index(level=0)
    feat=pd.DataFrame(feat)
    fin_res=pd.concat([fin_res,feat],axis=1)
    fin_res.drop(fin_res.loc[fin_res['P>|t|']==high].index, inplace=True)
    fin_res=fin_res.rename(columns={0:'feat'})
    feat=np.array(fin_res['feat'])
```
